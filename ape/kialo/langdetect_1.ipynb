```python
import os
import shutil
from langdetect import detect, LangDetectException

# Chemin du dossier contenant les fichiers de débats
base_folder_path = "../../data/kialo/debates/"

# Fonction pour classifier les fichiers par langue
def classify_files_by_language(base_folder_path):
    for filename in os.listdir(base_folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(base_folder_path, filename)
            
            # Lire le contenu du fichier
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            
            try:
                # Détecter la langue
                lang = detect(content)
                print(f"Le fichier '{filename}' est détecté comme étant en '{lang}'.")

                # Créer un dossier pour la langue détectée s'il n'existe pas
                lang_folder = os.path.join(base_folder_path, lang)
                if not os.path.exists(lang_folder):
                    os.makedirs(lang_folder)
                
                # Déplacer le fichier dans le dossier de la langue correspondante
                new_file_path = os.path.join(lang_folder, filename)
                shutil.copy(file_path, new_file_path)
            
            except LangDetectException:
                print(f"Impossible de détecter la langue du fichier '{filename}'.")

# Exécuter la fonction
classify_files_by_language(base_folder_path)
```
